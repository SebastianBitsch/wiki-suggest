```python
articles = read_articles_file('/work3/s204163/wiki/article_texts',100)
```

```python
import re
```

```python
def remove_numbers(string:str):
    pattern = r'[0-9]'

    # Match all digits in the string and replace them with an empty string
    new_string = re.sub(pattern, '', string)
    return new_string
```

```python
keys = list(articles.keys())
```

```python
corpus = [remove_numbers( articles[keys[i]][1]).lower() for i in range(len(articles))]
```

### Removing stopwords and Stemming the corpus

```python
stop_words = set(stopwords.words('english'))
```

```python
filtered_words = []
for text in corpus:
    for word in text:
        # filtered_words = [word for word in text if word not in stop_words]
        if word not in stop_words:
            filtered_words.append(word)
print(filtered_words)
```

```python
filtered_words
```

```python
filtered_corpus = []
stemmer = PorterStemmer()
for text in corpus:
    filtered_words = [word for word in text if word not in stop_words]
    stemmed_tokens = [stemmer.stem(word) for word in filtered_words]
    filtered_corpus.append(" ".join(stemmed_tokens))
```

```python

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(filtered_corpus)
vectorizer.get_feature_names_out()
print(X.shape)
```

```python
kmeans = KMeans(n_clusters=15, random_state=0, n_init="auto").fit(X)
kmeans.labels_
```

```python
desired_cluster =   0

# Filter and print the texts belonging to the desired cluster
texts_in_desired_cluster = [text for i, text in enumerate(corpus) if kmeans.labels_[i] == desired_cluster]

# Print texts
for text in texts_in_desired_cluster:
    print(text[:200])
    print('\n \n \n')
```
